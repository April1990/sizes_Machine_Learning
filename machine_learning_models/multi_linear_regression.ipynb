```python
import pandas as pd
import numpy as np
dataset = pd.read_csv(r'../csv/modified_male_female.csv')
dataset.loc[dataset['gender'] == 'Male', 'gender'] = 0
dataset.loc[dataset['gender'] == 'Female', 'gender'] = 1

dataset.head(3)
dataset.tail(3)
```

```python
import seaborn as sns
%matplotlib inline

sns.pairplot(dataset, hue='size')
```
